# Приведение столбцов к стилю camel_case

```python
# Импортируем библиотеку pandas
import pandas as pd

# Полный путь к файлу
file_path = '/content/drive/MyDrive/Portfolio/PN14/amazon_market_data.csv'

# Шаг 1: Загружаем датафрейм
df = pd.read_csv(file_path)

# Отображаем оригинальные названия столбцов
print("Оригинальные названия столбцов:")
print(df.columns.tolist())

# Шаг 2: Функция для преобразования в camelCase
def to_camel_case(text):
    # Разделяем по пробелам и другим возможным разделителям (можно расширить)
    words = text.split(' ')
    # Первое слово — в нижнем регистре, остальные — с заглавной буквы
    return words[0].lower() + ''.join(word.capitalize() for word in words[1:])

# Применяем преобразование ко всем названиям столбцов
df.columns = [to_camel_case(col) for col in df.columns]

# Шаг 3: Проверяем результат
print("\nНовые названия столбцов (camelCase):")
print(df.columns.tolist())

# Шаг 4 (опционально): Сохраняем обновлённый датафрейм
output_path = '/content/drive/MyDrive/Portfolio/PN14/amazon_market_data_camelcase.csv'
df.to_csv(output_path, index=False)

print(f"\nФайл сохранён по пути: {output_path}")
```

# Приведение типов данных

```python
# Импортируем библиотеку pandas
import pandas as pd

# Полный путь к файлу
file_path = '/content/drive/MyDrive/Portfolio/PN14/amazon_market_data.csv'

# Шаг 1: Загружаем датафрейм
df = pd.read_csv(file_path)

print("Первые 5 строк датафрейма (до преобразования):")
print(df[['Order Date']].head())

# Шаг 2: Преобразуем столбец 'Order Date' в тип datetime
# Указываем формат, так как дата в формате DD-MM-YY (например, 01-01-20)
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%y', errors='coerce')

# Пояснение:
# - format='%d-%m-%y': указывает, что дата в формате день-месяц-год (2 цифры года)
# - errors='coerce': преобразует некорректные значения в NaT (Not a Time), если есть

print("\nТип данных столбца 'Order Date' после преобразования:")
print(df['Order Date'].dtype)

print("\nПервые 5 строк после преобразования даты:")
print(df[['Order Date']].head())

# Шаг 3 (опционально): Сортируем по дате, чтобы найти начало и конец периода
df = df.sort_values('Order Date').reset_index(drop=True)

print(f"\n Минимальная дата (начало продаж): {df['Order Date'].min().strftime('%Y-%m-%d')}")
print(f" Максимальная дата (конец продаж): {df['Order Date'].max().strftime('%Y-%m-%d')}")

# Шаг 4 (опционально): Сохраняем обновлённый датафрейм
output_path = '/content/drive/MyDrive/Portfolio/PN14/amazon_market_data_cleaned.csv'
df.to_csv(output_path, index=False)
print(f"\n Данные сохранены в: {output_path}")
```

Вопрос 4. Изучить динамику количества заказов по дате заказов (по месяцам)

```python
# Импорт необходимых библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import MonthLocator, DateFormatter

# Настройка стиля
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 10
sns.set_style("whitegrid")

# Цвета бренда Amazon
AMAZON_ORANGE = "#FF9900"
AMAZON_DARK_GRAY = "#131A22"
AMAZON_BLUE_GRAY = "#232F3E"

# Путь к файлу
file_path = '/content/drive/MyDrive/Portfolio/PN14/amazon_market_data.csv'

# Шаг 1: Загрузка данных
df = pd.read_csv(file_path)
print(" Данные загружены. Размер датафрейма:", df.shape)

# Шаг 2: Преобразование столбца 'Order Date' в datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%y', errors='coerce')

# Проверка на некорректные даты
if df['Order Date'].isnull().any():
    print(f"  Обнаружено {df['Order Date'].isnull().sum()} некорректных дат. Удалены.")
    df = df.dropna(subset=['Order Date'])

# Сортируем по дате
df = df.sort_values('Order Date').reset_index(drop=True)

# Шаг 3: Извлечение месяца и года для группировки
df['Order Month'] = df['Order Date'].dt.to_period('M')  # Например, '2020-01'
df['Order Month Start'] = df['Order Month'].dt.start_time  # Для построения графика

print(f" Диапазон данных: с {df['Order Date'].min().strftime('%d.%m.%Y')} по {df['Order Date'].max().strftime('%d.%m.%Y')}")

# Шаг 4: Подсчёт количества уникальных заказов по месяцам
orders_by_month = df.groupby('Order Month Start')['Order ID'].nunique().reset_index()
orders_by_month.columns = ['Month', 'Number of Orders']

print(f"\n Количество заказов по месяцам:")
print(orders_by_month.to_string(index=False))

# Шаг 5: Визуализация динамики заказов
fig, ax = plt.subplots(figsize=(14, 6))

# Построение линейного графика
ax.plot(orders_by_month['Month'], orders_by_month['Number of Orders'],
        color=AMAZON_ORANGE, linewidth=3, marker='o', markersize=6, label='Количество заказов')

# Настройка фона
ax.set_facecolor('#f8f9fa')
fig.patch.set_facecolor('white')

# Настройка осей
ax.set_title(" Динамика количества заказов по месяцам", fontsize=16, fontweight='bold', color=AMAZON_DARK_GRAY)
ax.set_xlabel("Месяц", fontsize=12, color=AMAZON_BLUE_GRAY)
ax.set_ylabel("Количество заказов", fontsize=12, color=AMAZON_BLUE_GRAY)

# Форматирование оси X
ax.xaxis.set_major_locator(MonthLocator())
ax.xaxis.set_major_formatter(DateFormatter("%b %Y"))
plt.xticks(rotation=0, color=AMAZON_DARK_GRAY)
plt.yticks(color=AMAZON_DARK_GRAY)

# Сетка и легенда
ax.grid(True, which='major', axis='y', linestyle='--', alpha=0.5)
ax.legend(frameon=False, fontsize=10)

# Добавим значения над точками
for idx, row in orders_by_month.iterrows():
    ax.text(row['Month'], row['Number of Orders'] + 5,
            f"{row['Number of Orders']}", ha='center', va='bottom',
            fontsize=9, color=AMAZON_DARK_GRAY, fontweight='bold')

plt.tight_layout()
plt.show()


# Основные метрики
total_orders = orders_by_month['Number of Orders'].sum()
avg_monthly_orders = orders_by_month['Number of Orders'].mean()
max_month = orders_by_month.loc[orders_by_month['Number of Orders'].idxmax()]
min_month = orders_by_month.loc[orders_by_month['Number of Orders'].idxmin()]

print(f" Общее количество заказов за период: {total_orders:,}")
print(f" Среднее количество заказов в месяц: {avg_monthly_orders:.0f}")
print(f" Самый активный месяц: {max_month['Month'].strftime('%B %Y')} — {max_month['Number of Orders']} заказов")
print(f" Наименее активный месяц: {min_month['Month'].strftime('%B %Y')} — {min_month['Number of Orders']} заказов")

# Тренды
orders_by_month_sorted = orders_by_month.sort_values('Month')
first_half_avg = orders_by_month_sorted.iloc[:6]['Number of Orders'].mean()
second_half_avg = orders_by_month_sorted.iloc[6:]['Number of Orders'].mean()

print(f"\n Сравнение первой и второй половины периода:")
print(f"  • Среднее в 1-й половине (янв–июн): {first_half_avg:.0f} заказов/месяц")
print(f"  • Среднее во 2-й половине (июл–дек): {second_half_avg:.0f} заказов/месяц")

if second_half_avg > first_half_avg:
    print(" Наблюдается рост активности во второй половине года.")
else:
    print(" Активность снизилась во второй половине года.")

# Пиковые месяцы
peak_months = orders_by_month[orders_by_month['Number of Orders'] > avg_monthly_orders * 1.2]
print(f"\n Пиковые месяцы (на 20% выше среднего):")
for _, row in peak_months.iterrows():
    print(f"  • {row['Month'].strftime('%B %Y')}: {row['Number of Orders']} заказов")

# Сезонность
print(f"\n Возможная сезонность:")
if 'December' in max_month['Month'].strftime('%B'):
    print(" Рост в декабре — вероятно, новогодние/рождественские покупки.")
if 'July' in min_month['Month'].strftime('%B'):
    print(" Спад в июле — возможен летний спад активности.")
```

Анализ и закономерности:
1. Чёткая временная динамика - данные охватывают 12 месяцев (январь–декабрь), что позволяет анализировать годовые тренды.

2. Рост активности во второй половине года - заказы постепенно увеличиваются, особенно в осенние и зимние месяцы.

3. Пик активности - декабрь - это типичная сезонная закономерность: предновогодние закупки, подарки, распродажи.

4. Спад в июне-июле - возможно, связан с летним периодом, когда снижается деловая активность (особенно в сегменте Corporate и Home Office).

```python
# Импорт необходимых библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import MonthLocator, DateFormatter

# Настройка стиля графиков
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 10
sns.set_style("whitegrid")

# Цвета бренда Amazon
AMAZON_ORANGE = "#FF9900"     # Основной акцент
AMAZON_DARK_GRAY = "#131A22"  # Текст, заголовки
AMAZON_BLUE_GRAY = "#232F3E"  # Оси, легенда

# Путь к файлу
file_path = '/content/drive/MyDrive/Portfolio/PN14/amazon_market_data.csv'

# Шаг 1: Загрузка данных
df = pd.read_csv(file_path)
print(" Данные загружены. Размер датафрейма:", df.shape)

# Шаг 2: Преобразование 'Order Date' в datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%y', errors='coerce')

# Удаление строк с некорректной датой
df = df.dropna(subset=['Order Date']).reset_index(drop=True)

# Сортировка по дате
df = df.sort_values('Order Date')

# Извлечение месяца для группировки
df['Order Month'] = df['Order Date'].dt.to_period('M')
df['Month Start'] = df['Order Month'].dt.start_time

print(f" Диапазон данных: с {df['Order Date'].min().strftime('%d.%m.%Y')} по {df['Order Date'].max().strftime('%d.%m.%Y')}")

# Шаг 3: Группировка по месяцам — суммарные продажи
sales_by_month = df.groupby('Month Start')['Sales'].sum().reset_index()
sales_by_month.columns = ['Month', 'Total Sales']

print(f"\n Продажи по месяцам:")
print(sales_by_month.to_string(index=False, formatters={'Total Sales': '${:,.2f}'.format}))

# Шаг 4: Визуализация динамики продаж
fig, ax = plt.subplots(figsize=(14, 6))

# Линейный график
ax.plot(sales_by_month['Month'], sales_by_month['Total Sales'],
        color=AMAZON_ORANGE, linewidth=3, marker='o', markersize=6, label='Выручка (Sales)')

# Настройка фона
ax.set_facecolor('#f9f9f9')
fig.patch.set_facecolor('white')

# Заголовок и подписи
ax.set_title(" Динамика суммарных продаж по месяцам", fontsize=16, fontweight='bold', color=AMAZON_DARK_GRAY)
ax.set_xlabel("Месяц", fontsize=12, color=AMAZON_BLUE_GRAY)
ax.set_ylabel("Суммарные продажи ($)", fontsize=12, color=AMAZON_BLUE_GRAY)

# Форматирование оси X
ax.xaxis.set_major_locator(MonthLocator())
ax.xaxis.set_major_formatter(DateFormatter("%b %Y"))
plt.xticks(rotation=0, color=AMAZON_DARK_GRAY)
plt.yticks(color=AMAZON_DARK_GRAY)

# Сетка и легенда
ax.grid(False, axis='y', linestyle=' ', alpha=0.5)
ax.legend(frameon=False, fontsize=10)

# Добавление значений над точками
for idx, row in sales_by_month.iterrows():
    ax.text(row['Month'], row['Total Sales'] + 5000,
            f"${row['Total Sales']:,.0f}", ha='center', va='bottom',
            fontsize=9, color=AMAZON_DARK_GRAY, fontweight='bold')

plt.tight_layout()
plt.show()
```

Анализ и закономерности:
1. Чёткая сезонная динамика - продажи растут во второй половине года, достигая пика в декабре.

2. Декабрь - рекордный месяц - вероятно, связан с рождественскими и новогодними покупками, распродажами Black Friday и Cyber Monday.

3. Спад в июне-июле - типичный летний спад, особенно в сегменте Corporate, когда снижается деловая активность.

4. Рост с сентября по декабрь - показывает эффективность маркетинговых кампаний и повышенный спрос на офисные и технологические товары перед новым годом.

5. Высокая волатильность - разница между лучшим и худшим месяцем может превышать 2x, что указывает на необходимость управления запасами и логистикой.

```python
# Импорт библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Настройка стиля
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 10
sns.set_style("white")  # Полностью убирает сетку и фон

# Цвета бренда Amazon
AMAZON_ORANGE = "#FF9900"
AMAZON_DARK_GRAY = "#131A22"
AMAZON_BLUE_GRAY = "#232F3E"

# Путь к файлу
file_path = '/content/drive/MyDrive/Portfolio/PN14/amazon_market_data.csv'

# Шаг 1: Загрузка данных
df = pd.read_csv(file_path)
print(" Данные загружены. Размер датафрейма:", df.shape)

# Шаг 2: Преобразование 'Order Date' в datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%y', errors='coerce')

# Удаляем строки с некорректной датой
df = df.dropna(subset=['Order Date']).reset_index(drop=True)

# ===================================================================
# 1. СОЗДАНИЕ СВОДНОЙ ТАБЛИЦЫ ПО КАТЕГОРИЯМ И ПОДКАТЕГОРИЯМ
# ===================================================================

# Группировка по Category и Sub-Category, суммируем Sales
sales_by_category = df.groupby(['Category', 'Sub-Category'])['Sales'].sum().reset_index()
sales_by_category = sales_by_category.sort_values('Sales', ascending=False)

print("\n Продажи по категориям и подкатегориям (топ-10):")
print(sales_by_category.head(10).to_string(index=False, formatters={'Sales': '${:,.2f}'.format}))

# Также получим агрегацию только по Category
sales_by_main_category = df.groupby('Category')['Sales'].sum().reset_index()
sales_by_main_category = sales_by_main_category.sort_values('Sales', ascending=False)

print(f"\n Общие продажи по основным категориям:")
print(sales_by_main_category.to_string(index=False, formatters={'Sales': '${:,.2f}'.format}))

# ===================================================================
# 2. ВИЗУАЛИЗАЦИЯ — ПРОДАЖИ ПО ОСНОВНЫМ КАТЕГОРИЯМ
# ===================================================================

fig, ax = plt.subplots(1, 2, figsize=(18, 7))

# График 1: Продажи по Category (основные категории)
bars1 = ax[0].bar(sales_by_main_category['Category'],
                  sales_by_main_category['Sales'],
                  color=AMAZON_ORANGE, edgecolor=AMAZON_DARK_GRAY, linewidth=1)

ax[0].set_title("Продажи по категориям товаров", fontsize=14, fontweight='bold', color=AMAZON_DARK_GRAY)
ax[0].set_ylabel("Суммарные продажи ($)", fontsize=12, color=AMAZON_BLUE_GRAY)
ax[0].tick_params(axis='x', rotation=0, colors=AMAZON_DARK_GRAY)
ax[0].tick_params(axis='y', colors=AMAZON_DARK_GRAY)
ax[0].set_facecolor('white')  # Чистый белый фон

# Исправление ошибки: используем max() по всей колонке, а не по скаляру
max_sales = sales_by_main_category['Sales'].max()

# Подписи на столбцах
for bar in bars1:
    height = bar.get_height()
    ax[0].text(bar.get_x() + bar.get_width()/2., height + max_sales * 0.01,
               f'${height:,.0f}', ha='center', va='bottom', fontsize=10, color=AMAZON_DARK_GRAY)

# ===================================================================
# 3. ВИЗУАЛИЗАЦИЯ — ПРОДАЖИ ПО TOP-10 SUB-CATEGORIES
# ===================================================================

top_10_sub = sales_by_category.head(10)

bars2 = ax[1].barh(top_10_sub['Sub-Category'][::-1], top_10_sub['Sales'][::-1],
                   color=AMAZON_BLUE_GRAY, edgecolor=AMAZON_DARK_GRAY, linewidth=1)

ax[1].set_title("Топ-10 подкатегорий по продажам", fontsize=14, fontweight='bold', color=AMAZON_DARK_GRAY)
ax[1].set_xlabel("Суммарные продажи ($)", fontsize=12, color=AMAZON_BLUE_GRAY)
ax[1].tick_params(axis='x', colors=AMAZON_DARK_GRAY)
ax[1].tick_params(axis='y', colors=AMAZON_DARK_GRAY)
ax[1].set_facecolor('white')

# Подписи на горизонтальных столбцах
for i, bar in enumerate(bars2):
    width = bar.get_width()
    ax[1].text(width + max_sales * 0.01, bar.get_y() + bar.get_height()/2,
               f'${width:,.0f}', ha='left', va='center', fontsize=10, color=AMAZON_DARK_GRAY)

# Убираем все линии сетки
for axis in ['top', 'right', 'left', 'bottom']:
    ax[0].spines[axis].set_visible(False)
    ax[1].spines[axis].set_visible(False)

plt.tight_layout()
plt.show()
```

Анализ и закономерности:
 1. Technology - лидер по продажам (около 40-45% от общей выручки), несмотря на меньшее количество заказов. Это объясняется высокой стоимостью единицы товара (принтеры, серверы, телефоны).

 2. Office Supplies - объёмный, но низкомаржинальный сегмент. Занимает второе место по выручке, но средний чек ниже. Включает канцелярию, бумагу, скрепки - дешёвые, но частые покупки.

 3. Furniture - третий по объёму, но с высокой волатильностью. Включает дорогие товары (конференц-столы, стулья), но продажи менее стабильны.

 4. Топ-3 подкатегории:

```python
# Импорт библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Настройка стиля
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 10
sns.set_style("white")  # Убирает сетку и фон

# Цвета бренда Amazon
AMAZON_ORANGE = "#FF9900"
AMAZON_DARK_GRAY = "#131A22"
AMAZON_BLUE_GRAY = "#232F3E"

# Путь к файлу
file_path = '/content/drive/MyDrive/Portfolio/PN14/amazon_market_data.csv'

# Шаг 1: Загрузка данных
df = pd.read_csv(file_path)
print(" Данные загружены. Размер датафрейма:", df.shape)

# Шаг 2: Преобразование 'Order Date' в datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%y', errors='coerce')

# Удаляем строки с некорректной датой
df = df.dropna(subset=['Order Date']).reset_index(drop=True)

# ===================================================================
# 1. АНАЛИЗ ПОПУЛЯРНОСТИ ТОВАРОВ
# ===================================================================

# Группировка по Product Name: количество строк = количество позиций в заказах
product_popularity = df.groupby('Product Name').size().reset_index(name='Order Count')
product_sales = df.groupby('Product Name')['Sales'].sum().reset_index()
product_sales = product_sales.rename(columns={'Sales': 'Total Sales'})

# Объединяем данные
product_summary = pd.merge(product_popularity, product_sales, on='Product Name')
product_summary = product_summary.sort_values('Order Count', ascending=False)

print(f"\n Топ-10 самых популярных товаров (по количеству заказов):")
print(product_summary.head(10).to_string(index=False, formatters={'Order Count': '{:,.0f}'.format, 'Total Sales': '${:,.2f}'.format}))

# ===================================================================
# 2. ТОП-10 ПО КОЛИЧЕСТВУ ЗАКАЗОВ
# ===================================================================

top_10_by_count = product_summary.head(10)

# ===================================================================
# 3. ТОП-10 ПО ВЫРУЧКЕ
# ===================================================================

top_10_by_sales = product_summary.sort_values('Total Sales', ascending=False).head(10)
print(f"\n Топ-10 товаров по выручке:")
print(top_10_by_sales[['Product Name', 'Total Sales']].to_string(index=False, formatters={'Total Sales': '${:,.2f}'.format}))

# ===================================================================
# 4. ВИЗУАЛИЗАЦИЯ
# ===================================================================

fig, ax = plt.subplots(1, 2, figsize=(18, 7))

# График 1: Топ-10 по количеству заказов
bars1 = ax[0].barh(top_10_by_count['Product Name'][::-1], top_10_by_count['Order Count'][::-1],
                   color=AMAZON_ORANGE, edgecolor=AMAZON_DARK_GRAY, linewidth=1)
ax[0].set_title("Топ-10 товаров по количеству заказов", fontsize=14, fontweight='bold', color=AMAZON_DARK_GRAY)
ax[0].set_xlabel("Количество заказов", fontsize=12, color=AMAZON_BLUE_GRAY)
ax[0].tick_params(axis='x', colors=AMAZON_DARK_GRAY)
ax[0].tick_params(axis='y', colors=AMAZON_DARK_GRAY)
ax[0].set_facecolor('white')

# Подписи на столбцах
for bar in bars1:
    width = bar.get_width()
    ax[0].text(width + 0.1, bar.get_y() + bar.get_height()/2,
               f'{int(width)}', ha='left', va='center', fontsize=9, color=AMAZON_DARK_GRAY)

# График 2: Топ-10 по выручке
bars2 = ax[1].barh(top_10_by_sales['Product Name'][::-1], top_10_by_sales['Total Sales'][::-1],
                   color=AMAZON_BLUE_GRAY, edgecolor=AMAZON_DARK_GRAY, linewidth=1)
ax[1].set_title("Топ-10 товаров по выручке", fontsize=14, fontweight='bold', color=AMAZON_DARK_GRAY)
ax[1].set_xlabel("Суммарные продажи ($)", fontsize=12, color=AMAZON_BLUE_GRAY)
ax[1].tick_params(axis='x', colors=AMAZON_DARK_GRAY)
ax[1].tick_params(axis='y', colors=AMAZON_DARK_GRAY)
ax[1].set_facecolor('white')

# Подписи на столбцах
for bar in bars2:
    width = bar.get_width()
    ax[1].text(width + max(top_10_by_sales['Total Sales'])*0.01, bar.get_y() + bar.get_height()/2,
               f'${width/1e3:.1f}K', ha='left', va='center', fontsize=9, color=AMAZON_DARK_GRAY)

# Убираем границы осей
for a in ax:
    for spine in a.spines.values():
        spine.set_visible(False)

plt.tight_layout()
plt.show()
```

   - Самый прибыльный - редкий, но дорогой.
4. Категории лидеров:
   - По количеству заказов: Office Supplies (особенно Binders, Supplies, Fasteners).
   - По выручке: Furniture и Technology.

```python
# Импорт библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Настройка стиля
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 10
sns.set_style("white")  # Убирает сетку и фон

# Цвета бренда Amazon
AMAZON_ORANGE = "#FF9900"
AMAZON_DARK_GRAY = "#131A22"
AMAZON_BLUE_GRAY = "#232F3E"

# Путь к файлу
file_path = '/content/drive/MyDrive/Portfolio/PN14/amazon_market_data.csv'

# Шаг 1: Загрузка данных
df = pd.read_csv(file_path)
print(" Данные загружены. Размер датафрейма:", df.shape)

# Шаг 2: Преобразование 'Order Date' в datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%y', errors='coerce')
df = df.dropna(subset=['Order Date']).reset_index(drop=True)

# ===================================================================
# 1. АНАЛИЗ РАСПРЕДЕЛЕНИЯ КАТЕГОРИЙ ПО ШТАТАМ
# ===================================================================

# Группировка по State и Category
state_category_sales = df.groupby(['State', 'Category'])['Sales'].sum().unstack(fill_value=0)

# Добавляем общую выручку по штатам для сортировки
state_category_sales['Total'] = state_category_sales.sum(axis=1)
state_category_sales = state_category_sales.sort_values('Total', ascending=False).drop('Total', axis=1)

print("\n Продажи по категориям в топ-10 штатов:")
print(state_category_sales.head(10).round(2))

# ===================================================================
# 2. АНАЛИЗ ПО ГОРОДАМ (Top-10 городов)
# ===================================================================

# Группировка по City и Category
city_category_sales = df.groupby(['City', 'Category'])['Sales'].sum().unstack(fill_value=0)

# Сортировка по общей выручке
city_category_sales['Total'] = city_category_sales.sum(axis=1)
city_category_sales = city_category_sales.sort_values('Total', ascending=False).drop('Total', axis=1)

print("\n Продажи по категориям в топ-10 городов:")
print(city_category_sales.head(10).round(2))

# ===================================================================
# 3. ВИЗУАЛИЗАЦИЯ
# ===================================================================

fig, ax = plt.subplots(2, 1, figsize=(16, 12))

# График 1: Распределение категорий по топ-10 штатам
top_10_states = state_category_sales.head(10)
top_10_states.plot(kind='bar', stacked=True, ax=ax[0],
                   color=[AMAZON_ORANGE, AMAZON_BLUE_GRAY, AMAZON_DARK_GRAY],
                   edgecolor=AMAZON_DARK_GRAY, linewidth=0.8)

ax[0].set_title("Распределение продаж по категориям в топ-10 штатов", fontsize=14, fontweight='bold', color=AMAZON_DARK_GRAY)
ax[0].set_ylabel("Продажи ($)", fontsize=12, color=AMAZON_DARK_GRAY)
ax[0].tick_params(axis='x', rotation=45, colors=AMAZON_DARK_GRAY)
ax[0].tick_params(axis='y', colors=AMAZON_DARK_GRAY)
ax[0].legend(title='Категория', edgecolor='white')
ax[0].set_facecolor('white')

# Убираем сетку
ax[0].grid(False)

# График 2: Распределение категорий по топ-10 городам
top_10_cities = city_category_sales.head(10)
top_10_cities.plot(kind='bar', stacked=True, ax=ax[1],
                   color=[AMAZON_ORANGE, AMAZON_BLUE_GRAY, AMAZON_DARK_GRAY],
                   edgecolor=AMAZON_DARK_GRAY, linewidth=0.8)

ax[1].set_title("Распределение продаж по категориям в топ-10 городах", fontsize=14, fontweight='bold', color=AMAZON_DARK_GRAY)
ax[1].set_ylabel("Продажи ($)", fontsize=12, color=AMAZON_DARK_GRAY)
ax[1].tick_params(axis='x', rotation=45, colors=AMAZON_DARK_GRAY)
ax[1].tick_params(axis='y', colors=AMAZON_DARK_GRAY)
ax[1].legend().set_visible(False)  # Убираем дублирование легенды
ax[1].set_facecolor('white')

# Убираем сетку
ax[1].grid(False)

plt.tight_layout()
plt.show()
```

Вопрос 10: Изучить динамику продаж внутри категорий

```python
# Импорт библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import MonthLocator, DateFormatter

# Настройка стиля
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 10
sns.set_style("white")  # Убирает сетку и фон

# Цвета бренда Amazon
AMAZON_ORANGE = "#FF9900"     # Technology
AMAZON_BLUE_GRAY = "#232F3E"  # Furniture
AMAZON_DARK_GRAY = "#131A22"  # Office Supplies

COLOR_MAP = {
    'Technology': AMAZON_ORANGE,
    'Furniture': AMAZON_BLUE_GRAY,
    'Office Supplies': AMAZON_DARK_GRAY
}

# Путь к файлу
file_path = '/content/drive/MyDrive/Portfolio/PN14/amazon_market_data.csv'

# Шаг 1: Загрузка данных
df = pd.read_csv(file_path)
print(" Данные загружены. Размер датафрейма:", df.shape)

# Шаг 2: Преобразование 'Order Date' в datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%y', errors='coerce')
df = df.dropna(subset=['Order Date']).reset_index(drop=True)

# Сортировка по дате
df = df.sort_values('Order Date')

# Извлечение месяца
df['Order Month'] = df['Order Date'].dt.to_period('M')
df['Month Start'] = df['Order Month'].dt.start_time

# Шаг 3: Группировка по месяцу и категории
monthly_sales_by_category = df.groupby(['Month Start', 'Category'])['Sales'].sum().reset_index()

# Убедимся, что все месяцы в диапазоне присутствуют
all_months = pd.period_range(df['Order Month'].min(), df['Order Month'].max(), freq='M').start_time
categories = df['Category'].unique()

# Восстанавливаем полный временной ряд
full_index = pd.MultiIndex.from_product([all_months, categories], names=['Month Start', 'Category'])
filled_data = monthly_sales_by_category.set_index(['Month Start', 'Category']).reindex(full_index, fill_value=0).reset_index()
filled_data['Sales'] = filled_data['Sales'].round(2)

print(f"\n Продажи по категориям и месяцам (пример):")
print(filled_data.head(10).to_string(index=False))

# Шаг 4: Визуализация динамики
fig, ax = plt.subplots(figsize=(14, 6))

for category in categories:
    data = filled_data[filled_data['Category'] == category]
    ax.plot(data['Month Start'], data['Sales'],
            label=category,
            color=COLOR_MAP.get(category, 'gray'),
            linewidth=3 if category == 'Technology' else 2,
            marker='o', markersize=5)

# Настройка графика
ax.set_title(" Динамика продаж по категориям", fontsize=16, fontweight='bold', color=AMAZON_DARK_GRAY)
ax.set_xlabel("Месяц", fontsize=12, color=AMAZON_DARK_GRAY)
ax.set_ylabel("Суммарные продажи ($)", fontsize=12, color=AMAZON_DARK_GRAY)

# Форматирование оси X
ax.xaxis.set_major_locator(MonthLocator())
ax.xaxis.set_major_formatter(DateFormatter("%b %Y"))
plt.xticks(rotation=0, color=AMAZON_DARK_GRAY)
plt.yticks(color=AMAZON_DARK_GRAY)

# Легенда
ax.legend(title='Категория', frameon=False, fontsize=11, title_fontsize=12)

# Фон и границы
ax.set_facecolor('white')
fig.patch.set_facecolor('white')

# Убираем сетку
ax.grid(False)

# Убираем лишние границы
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()

# ===================================================================
# 5. АНАЛИЗ И ЗАКОНОМЕРНОСТИ
# ===================================================================

print("\n" + "="*60)
print(" АНАЛИЗ ДИНАМИКИ ПРОДАЖ ПО КАТЕГОРИЯМ")
print("="*60)

# Общие метрики
total_sales = filled_data['Sales'].sum()
print(f" Общие продажи за период: ${total_sales:,.2f}")

# Анализ по категориям
for category in categories:
    cat_data = filled_data[filled_data['Category'] == category]
    total = cat_data['Sales'].sum()
    peak_month = cat_data.loc[cat_data['Sales'].idxmax()]
    min_month = cat_data.loc[cat_data['Sales'].idxmin()]

    print(f"\n Категория: {category}")
    print(f"  • Вклад в общий объём: {total / total_sales * 100:.1f}%")
    print(f"  • Пик продаж: {peak_month['Month Start'].strftime('%B %Y')} — ${peak_month['Sales']:,.2f}")
    print(f"  • Минимум: {min_month['Month Start'].strftime('%B %Y')} — ${min_month['Sales']:,.2f}")

    # Рост во второй половине года
    first_half = cat_data[cat_data['Month Start'].dt.month <= 6]['Sales'].mean()
    second_half = cat_data[cat_data['Month Start'].dt.month > 6]['Sales'].mean()

    if second_half > first_half:
        print(f"   Продажи росли во второй половине года (+{((second_half - first_half) / first_half * 100):.1f}%)")
    else:
        print(f"   Продажи снизились во второй половине года")
```

Анализ и закономерности:
1. Technology - самый волатильный сектор:
   - Ярко выраженный пик в декабре - вероятно, связан с праздничными покупками, Black Friday, Cyber Monday.
   - Техника (телефоны, серверы, аксессуары) - частый подарок и инвестиция в начало года.
   - Резкий рост с сентября по декабрь.
2. Furniture - стабильный, но сезонный:
   - Пик в октябре-ноябре - возможно, подготовка к зимнему сезону, переезды, обустройство домашних офисов.
   - Спад в июне-июле - летний период, снижение деловой активности.
   - Домашние офисы - ключевой драйвер.
3. Office Supplies - наименее волатильный:
   - Более равномерное распределение.
   - Небольшой рост в январе - начало финансового года, закупки для офисов.
   - Сезонность слабее, но всё же заметна в декабре.
4. Общая закономерность:
   - Все категории показывают рост во второй половине года.
   - Ноябрь-декабрь - пик для всех трёх категорий.
   - Июнь-июль - спад, особенно у Furniture и Technology.
